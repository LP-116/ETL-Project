```python
import pandas as pd
from config import *
from sqlalchemy import create_engine
```

```python
regional_growth_load = "Pop_growth_CSVs/Aust_Data_Regional_Growth.csv"
```

```python
regional_growth_data = pd.read_csv(regional_growth_load)
regional_growth_data
```

```python
regional_growth_data["Local Government Area"] = regional_growth_data["Local Government Area"].str.replace(r"\(.*\)","")
```

```python
regional_growth_data["Local Government Area"] = regional_growth_data["Local Government Area"].str.strip()
```

```python
regional_growth_data.head(50)
```

```python
regional_growth_data = regional_growth_data.rename(columns={"LGA code": "lga_code", "Local Government Area": "local_government_area",
                                    "2018": "pop_2018", "2019": "pop_2019", "Growth no.":"growth_no", "Growth %":"growth_per",
                                    "Natural increase no.": "births_no", "Net internal migration no.": "int_migration_no",
                                    "Net overseas migration no.":"os_migration_no", "Area (km2)":"lga_area", 
                                    "Population density 2019 (persons/km2)":"pop_density"})
regional_growth_data
```

```python
connection_string = "{}:{}@localhost:5432/LGA_db".format(username, password)
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
lga_cols = ["local_government_area"]
lga_transformed = regional_growth_data[lga_cols].copy()
# lga_transformed.set_index("local_government_area", inplace=True)
# duplicate = lga_transformed[lga_transformed.duplicated(["local_government_area"], keep=False)]
# duplicate

lga_transformed = lga_transformed.drop_duplicates()
lga_transformed
```

```python
lga_transformed.to_sql(name='local_gov_area', con=engine, if_exists='append', index=False)
```

```python
location_cols = ["lga_code", "local_government_area"]
location_transformed = regional_growth_data[location_cols].copy()
location_transformed.set_index("lga_code", inplace=True)
location_transformed
```

```python
location_transformed.to_sql(name='location_id', con=engine, if_exists='append', index=True)
```

```python
pop_growth_cols = ["lga_code", "growth_no", "growth_per"]
growth_transformed = regional_growth_data[pop_growth_cols].copy()
growth_transformed.index.name = "growth_index"
growth_transformed
```

```python
growth_transformed.to_sql(name='population_growth', con=engine, if_exists='append', index=True)
```

```python
pop_num_cols = ["lga_code", "pop_2018", "pop_2019"]
num_transformed = regional_growth_data[pop_num_cols].copy()
num_transformed.index.name = "pop_index"
num_transformed
```

```python
num_transformed.to_sql(name='population_numbers', con=engine, if_exists='append', index=True)
```

```python
pop_density_cols = ["lga_code", "lga_area", "pop_density"]
density_transformed = regional_growth_data[pop_density_cols].copy()
density_transformed.index.name = "density_index"
density_transformed
```

```python
density_transformed.to_sql(name='population_density', con=engine, if_exists='append', index=True)
```

```python
growth_type_cols = ["lga_code", "births_no", "int_migration_no", "os_migration_no"]
growth_type_transformed = regional_growth_data[growth_type_cols].copy()
growth_type_transformed.index.name = "growth_type_index"
growth_type_transformed
```

```python
growth_type_transformed.to_sql(name='pop_growth_type', con=engine, if_exists='append', index=True)
```
